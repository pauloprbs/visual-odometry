```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
cam0_imgs = pd.read_csv('../VO/MH_01_easy/mav0/cam0/data.csv')
cam1_imgs = pd.read_csv('../VO/MH_01_easy/mav0/cam1/data.csv')
cam0_path = '../VO/MH_01_easy/mav0/cam0/data/'
cam1_path = '../VO/MH_01_easy/mav0/cam1/data/'
pose_it = 0

fast = cv.FastFeatureDetector_create()
brief = cv.xfeatures2d.BriefDescriptorExtractor_create()
bf = cv.BFMatcher(cv.NORM_HAMMING, crossCheck=True)
```

```python
img0 = cv.imread(cam0_path + cam0_imgs.at[pose_it,'filename'])
img1 = cv.imread(cam1_path + cam1_imgs.at[pose_it,'filename'])

kp0 = fast.detect(img0,None)
kp1 = fast.detect(img1,None)

img0_kp = cv.drawKeypoints(img0, kp0, None, color=(255, 0, 0))
img1_kp = cv.drawKeypoints(img1, kp0, None, color=(255, 0, 0))

kp0, des0 = brief.compute(img0,kp0)
kp1, des1 = brief.compute(img1,kp1)

matches = bf.match(des0, des1)

best_matches = sorted(matches, key = lambda x:x.distance)

img_matches = cv.drawMatches(img0,kp0,img1,kp1,best_matches[:2000],None,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111)
ax.imshow(img_matches), plt.show()
```
